```python
hello("cat")

bye("dog")
```

```python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
        
    def hello(self):
        print("Hello " + self.name + " !")
        
    def goodbye(self):
        print("Good-bye " + self.name + " !")
```

```python
m = Man("David")
```

```python
m.hello()
```

```python
m.goodbye()
```
