```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
```

* 다운로드받기

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
print(mnist.train.images.shape) ##mnist는 dataset의 이름 / train; 내가 컴퓨터에게 주는 data
print(mnist.test.images.shape)
```

```python
print(mnist.train.labels.shape)
print(mnist.test.labels.shape)
```

```python
idx = 47
```

```python
img1 = mnist.test.images[idx]
img1 = np.array(img1, dtype='float')
pixels = img1.reshape((28, 28))
plt.imshow(pixels, cmap='gray')
plt.title('mnist.test.images[{}]'.format(idx))
plt.show()
```

```python
label1 = mnist.test.labels[idx]
print(label1)
```

```python
np.argmax(label1)
```

* 아래 코드의 의미는?

```python
label2 = np.argmax(label1)
print(label2)
```

하나 더 연습하기

```python
np.argmax([0, 1, 6, 3, 2, -2, 5, 7, 10]) ##제일 큰 것의 위치가 숫자로 나온다. list max는 가장 큰 것을 찾는 것이고, arg는 자리를 찾는 것이다.
```

```python
label3 = np.argmax(label1)
print(label3)
```

```python
idx = 87
img1 = mnist.test.images[idx]
img1 = np.array(img1, dtype='float')
lbl1 = np.argmax(mnist.test.labels[idx])

pixels = img1.reshape((28, 28))
plt.imshow(pixels, cmap='gray')
plt.title('mnist.test.images[{}] ---> {}'.format(idx, lbl1))
plt.show()
```
