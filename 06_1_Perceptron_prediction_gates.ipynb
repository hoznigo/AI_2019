### Perceptrons - Making Predictions
#### Creating a gate with Perceptron

```python
import numpy as np
```

### AND Gate

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) #입력에 대한 가중치
    b = -0.7
    #activation 계산
    tmp = w[0]*x[0] + w[1]*x[1] + b # tmp = np.sum(w*x) + b
    print(tmp)
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
def AND1(x1, x2):
  X1 = np.array(x1)
  X2 = np.array(x2)
  w1 = np.array(0.5)
  w2 = np.array(0.5)
  b = -0.7
  tmp = X1 * w1 + X2 * w2 + b
  print(tmp)
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
AND1(1,0)
```

```python
AND1(1,1)
```

```python
AND1(0,0)
```

```python
AND1(0,1)
```

### NAND Gate

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = w[0]*x[0] + w[1]*x[1] + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
NAND(0,0)
```

```python
NAND(1,0)
```

```python
NAND(1,1)
```

### OR Gate

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0,0)
```

```python
OR(1,0)
```

```python
OR(0,1)
```

```python
OR(1,1)
```

### XOR Gate

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(0,0)
```

```python
XOR(1,1)
```

```python
XOR(0,1)
```

```python
XOR(1,0)
```

XOR cannot be expressed as a single layer Perceptron.
